```python
import pandas as pd
import numpy as np
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder 
class DataPrepKit:

    def __init__(self):
        pass

    # Data Reading
    def read_data(self):
        file_path = input("Enter file path: ")
        
        if 'csv' in file_path:
            df = pd.read_csv(file_path)
        elif 'excel' in file_path:
            df = pd.read_excel(file_path)
        elif 'json' in file_path:
            df = pd.read_json(file_path)
        
        return df
    
    # Data Summaries
    def summary(self, df):
        return df.describe()
    
    #Handling Missing Values
    def handle_missing_values(self, df, strategy):
        if strategy == 'drop':
            clean_df = df.dropna()
        elif strategy in ['mean', 'median', 'mode']:
            clean_df = df.fillna(df.agg(strategy))
        elif strategy in ['ffill', 'bfill']:
            clean_df = df.fillna(method=strategy)
        else:
            raise ValueError("Invalid strategy. Choose from 'mean', 'median', 'mode', 'ffill', 'bfill', 'drop'.")
    
        return clean_df

    # Data Encoding
    def encode_categorical(self, data, by, strategy):
        if strategy == 'one-hot':
            data_passthrough = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [by])], remainder='passthrough')
            data_encoded = np.array(data_passthrough.fit_transform(data))
        elif strategy == 'label':
            data_encoded = LabelEncoder().fit_transform(data[by])
        else:
            raise ValueError("Invalid strategy. Choose from 'one-hot' or 'label'.")

        return data_encoded
    
```

```python
#Usage example
dpk = DataPrepKit()

df = dpk.read_data()
```

```python
dpk.encode_categorical(df, 'Company', 'label')
```
